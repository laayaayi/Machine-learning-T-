```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data
```

```python
theta = np.pi / 6

A = np.array([
[np.cos(theta), -np.sin(theta)],
[np.sin(theta), np.cos(theta)]
])
```

```python
X_2d = X[:, :2].T
X_transformed = A @ X_2d
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].scatter(X_2d[0], X_2d[1], c=iris.target, cmap='viridis')
axs[0].set_title("Original (first 2 features)")
axs[0].axis('equal')
axs[0].grid(True)

axs[1].scatter(X_transformed[0], X_transformed[1], c=iris.target, cmap='viridis')
axs[1].set_title("After Linear Transform (Rotation)")
axs[1].axis('equal')
axs[1].grid(True)


plt.tight_layout()
plt.show()
```
